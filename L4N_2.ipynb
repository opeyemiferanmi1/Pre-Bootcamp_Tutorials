```
from sklearn.metrics import roc_auc_score
def model_auc(model):
    train_auc = roc_auc_score(ytrain, model.predict_proba(xtrain)[:, 1])
    val_auc = roc_auc_score(yval, model.predict_proba(xval)[:, 1])
    print(f'Train AUC: {train_auc}, Val Auc: {val_auc}')
```

```
from sklearn.model_selection import train_test_split
xtrain, xval, ytrain, yval = train_test_split(X, y, random_state=42, test_size=0.2, stratify=y)
```

```
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=700)
clf.fit(xtrain,ytrain)
y_pred = clf.predict(xval)
model_auc(clf)#83660
```

```
#THIS TOO
#from sklearn.ensemble import RandomForestClassifier
#from catboost import Pool, CatBoostClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import MinMaxScaler 
from sklearn.pipeline import Pipeline
import lightgbm as lgb
from sklearn.ensemble import RandomForestClassifier
from catboost import CatBoostClassifier, Pool, cv
from rgf.sklearn import RGFClassifier
from sklearn.ensemble import VotingClassifier
strKfold = StratifiedKFold(n_splits=25,shuffle=True)
#import lightgbm as lgb
cat1 = CatBoostClassifier(depth=6,iterations=1000, learning_rate=0.03,l2_leaf_reg = 7,eval_metric='AUC', random_seed=42,verbose = False)#846
#cat2 = CatBoostClassifier(thread_count =-1,rsm = 0.98,eval_metric='AUC',verbose=False)#43
cat3 = CatBoostClassifier(eval_metric='AUC',verbose=False)#845
#lgb1 = lgb.LGBMClassifier(n_jobs =-1,verbose =0,learning_rate=0.01,n_estimators=1360,max_depth=5)
lgb2 = lgb.LGBMClassifier(colsample_bytree = 0.53188948770158,verbose=2, learning_rate = 0.1,
                          max_depth= 21, min_child_samples = 313, min_child_weight= 0.01,
                          num_leaves = 25, reg_alpha=2, reg_lambda = 0,
                          subsample = 0.9373388595945467)#8443
from rgf.sklearn import RGFClassifier
rgf = RGFClassifier(l2= 0.1, max_leaf= 1000, min_samples_leaf= 5)
#rgf = RGFClassifier(l2= 0.1, max_leaf= 1000, min_samples_leaf= 5)#8442
rf = RandomForestClassifier(n_estimators = 700,max_features = 'sqrt', max_depth = 8, criterion ='gini', random_state=42)
```

```
#This one
 
scores = []
preds = []
#catboost_model1 = CatBoostClassifier(eval_metric='AUC', random_seed=12345,verbose = False)
model = VotingClassifier(estimators=[('CatBoost1', cat1),('CatBoost3', cat3),('LGB',lgb2),('RFC',rf)], voting='soft',weights=[3,2,1,1])
i = 1
for train, val in strKfold.split(X,y):
    pipe = Pipeline([('scaler', MinMaxScaler()), ('model', model)])
    xtrain, xval, ytrain, yval = X.iloc[train], X.iloc[val], y.iloc[train], y.iloc[val]
    #cat_features = cate_features_index
    pipe.fit(xtrain, ytrain)
    result = model_auc(pipe)
    score = roc_auc_score (yval, pipe.predict_proba(xval)[:, 1])
    pred = pipe.predict_proba(test_new)[:,1]
    scores.append(score)
    preds.append(pred)
    i += 1
 
test_pred = np.mean(scores)
pred_result = np.mean(preds)
```
